```python
import cv2
import matplotlib.pyplot as plt
```

```python
#resimleri yukleme ve renklerini duzeltme
img1 = cv2.imread("ronaldo.jpg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread("messi.jpg")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

plt.figure()
plt.imshow(img1)

plt.figure()
plt.imshow(img2)
```

```python
#simdi bunlarn boylarna bakalm. boylarnn ayn olmas lazm cunku farkl boyutlardaki matrisler toplanamaz. ayn yapcaz.
print(img1.shape)
print(img2.shape)
```

```python
#yeniden boyutlandryoruz.
img1 = cv2.resize(img1, (600,600))
print(img1.shape)

img2 = cv2.resize(img2, (600,600))
print(img2.shape)
```

```python
#fotograflar getirelim
plt.figure()
plt.imshow(img1)

plt.figure()
plt.imshow(img2)
```

```python
#simdi resimleri karstryoruz bu ise syle oluyor: alpha*img1 + beta*img2  seklinde
blended = cv2.addWeighted(src1 = img1, alpha = 0.5, src2 = img2, beta = 0.5, gamma = 0)
plt.figure()
plt.imshow(blended)
```
